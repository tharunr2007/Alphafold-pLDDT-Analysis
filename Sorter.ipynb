```python
import os
import send2trash

# Define the directory path
folder_path = r"D:\Research\Part 1 - Comparison\Comparison\E.Coli\AF 3 Prediction\Batch 1"

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    # Check if it's a file and not a directory
    file_path = os.path.join(folder_path, filename)

    # Only delete files that don't contain "model_1" in the filename
    if os.path.isfile(file_path) and "model_1" not in filename:
        # Move the file to the recycle bin
        send2trash.send2trash(file_path)

print("Files without 'model_1' in the name have been moved to the Recycle Bin.")
```

Move files without "model_1" to the trash as all the files are essentially the same and it leads to easier organization (each individual file will have a model_1 element)
