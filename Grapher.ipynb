```python
import os  # Imports the os module for interacting with the operating system's file system
import pandas as pd  # Imports the pandas module for data manipulation and analysis
import numpy as np  # Imports the numpy module for numerical operations
import matplotlib.pyplot as plt  # Imports the matplotlib module for plotting graphs
from Bio.PDB import PDBParser  # Imports the PDBParser class from Biopython for parsing PDB files

def extract_protein_data(pdb_file):
    """Extracts protein length and average pLDDT (B-factor) from a PDB file."""
    parser = PDBParser(QUIET=True)  # Initializes the PDB parser in quiet mode to suppress warnings
    structure = parser.get_structure('protein', pdb_file)  # Parses the PDB file to get the structure

    residue_indices = set()  # Initializes a set to store unique residue indices
    beta_factors = []  # Initializes a list to store B-factors (pLDDT scores)

    for model in structure:  # Iterates over each model in the structure
        for chain in model:  # Iterates over each chain in the model
            for residue in chain:  # Iterates over each residue in the chain
                residue_index = residue.id[1]  # Extracts the residue index
                residue_indices.add(residue_index)  # Adds the residue index to the set
                for atom in residue:  # Iterates over each atom in the residue
                    beta_factors.append(atom.get_bfactor())  # Appends the B-factor of the atom to the list

    protein_length = len(residue_indices)  # Calculates the length of the protein (number of unique residues)
    average_plddt = np.mean(beta_factors)  # Calculates the average pLDDT (mean of B-factors)

    return protein_length, average_plddt  # Returns the protein length and average pLDDT

def plot_protein_data(protein_lengths, average_plddts, output_file):
    """Plots protein length vs. average pLDDT and saves the plot to a file."""
    df = pd.DataFrame({'Protein Length': protein_lengths, 'Average pLDDT': average_plddts})  # Creates a DataFrame from the protein data

    plt.figure(figsize=(10, 6))  # Creates a new figure for the plot with a specified size
    plt.scatter(df['Protein Length'], df['Average pLDDT'], marker='o')  # Creates a scatter plot of protein length vs. average pLDDT
    plt.title('Protein Length vs Average pLDDT')  # Sets the title of the plot
    plt.xlabel('Protein Length (Number of Residues)')  # Sets the label for the x-axis
    plt.ylabel('Average pLDDT')  # Sets the label for the y-axis
    plt.grid(True)  # Enables grid lines on the plot
    plt.savefig(output_file)  # Saves the plot to the specified output file
    plt.show()  # Displays the plot

def main(pdb_dir, output_file):
    """Main function to process all PDB files in a directory and plot the results."""
    pdb_files = [os.path.join(pdb_dir, f) for f in os.listdir(pdb_dir) if f.endswith('.pdb')]  # Lists all PDB files in the specified directory

    protein_lengths = []  # Initializes a list to store protein lengths
    average_plddts = []  # Initializes a list to store average pLDDT values

    for pdb_file in pdb_files:  # Iterates over each PDB file
        protein_length, average_plddt = extract_protein_data(pdb_file)  # Extracts protein data from the PDB file
        protein_lengths.append(protein_length)  # Appends the protein length to the list
        average_plddts.append(average_plddt)  # Appends the average pLDDT to the list

    plot_protein_data(protein_lengths, average_plddts, output_file)  # Plots the protein data and saves the plot

if __name__ == "__main__":
    pdb_dir = r'D:\Research\Part 1 - Comparison\Scerevisiae\dataset'  # Specifies the directory containing the PDB files
    output_file = r'D:\Research\Graph Output\protein_length_vs_plddt_S.Scerevisiae.png'  # Specifies the path for the output plot file
    main(pdb_dir, output_file)  # Calls the main function with the specified PDB directory and output file
```

This script processes a directory of Protein Data Bank (PDB) files to analyze protein structures. For each PDB file, it calculates the protein length (number of unique residues) and the average pLDDT (predicted Local Distance Difference Test, stored in the B-factor field of the PDB file). It then visualizes the results in a scatter plot of protein length versus average pLDDT, which is saved to a specified file.
